# 1.LeetCode 225. Implement Stack using Queues

```python
from collections import deque

class MyStack:

    def __init__(self):
        self.q = deque()

    def push(self, x: int) -> None:
        self.q.append(x)
        # 큐의 원소를 한 칸씩 앞으로 보내 마지막에 들어온 원소가 맨 앞에 있도록 회전
        for _ in range(len(self.q) - 1):
            self.q.append(self.q.popleft())

    def pop(self) -> int:
        return self.q.popleft()

    def top(self) -> int:
        return self.q[0]

    def empty(self) -> bool:
        return not self.q
```

# LeetCode 232. Implement Queue using Stacks

```python
class MyQueue:

    def __init__(self):
        self.in_stack = []
        self.out_stack = []

    def push(self, x: int) -> None:
        self.in_stack.append(x)

    def pop(self) -> int:
        self.peek()  # out_stack에 값이 없으면 옮겨줌
        return self.out_stack.pop()

    def peek(self) -> int:
        if not self.out_stack:
            while self.in_stack:
                self.out_stack.append(self.in_stack.pop())
        return self.out_stack[-1]

    def empty(self) -> bool:
        return not self.in_stack and not self.out_stack
```

```python
## 3.02
from collections import deque

def is_in_language(s):
    if s.count('$') != 1:
        return False

    left, right = s.split('$')

    q = deque()

    for ch in left:
        q.append(ch)

    for ch in reversed(right):
        if not q or q.popleft() != ch:
            return False

    return not q
```

```python
## 3.04

from collections import deque

class StackUsingQueues:
    def __init__(self):
        self.q1 = deque()
        self.q2 = deque()

    def push(self, x):
        self.q1.append(x)  # 그냥 q1에 넣음

    def pop(self):
        if not self.q1:
            raise IndexError("pop from empty stack")

        # q1에 하나만 남을 때까지 q2로 옮김
        while len(self.q1) > 1:
            self.q2.append(self.q1.popleft())

        # 마지막 하나가 스택에서 꺼낼 값
        popped = self.q1.popleft()

        # q1과 q2를 바꿈 (q1이 항상 주요 큐)
        self.q1, self.q2 = self.q2, self.q1

        return popped

    def top(self):
        if not self.q1:
            raise IndexError("top from empty stack")

        while len(self.q1) > 1:
            self.q2.append(self.q1.popleft())

        top_elem = self.q1.popleft()
        self.q2.append(top_elem)

        self.q1, self.q2 = self.q2, self.q1
        return top_elem

    def is_empty(self):
        return not self.q1
```

```python
## 3.05

class QueueUsingStacks:
    def __init__(self):
        self.stack_in = []
        self.stack_out = []

    def enqueue(self, x):
        self.stack_in.append(x)

    def dequeue(self):
        if not self.stack_out:
            if not self.stack_in:
                raise IndexError("dequeue from empty queue")
            # stack_in -> stack_out
            while self.stack_in:
                self.stack_out.append(self.stack_in.pop())

        return self.stack_out.pop()

    def peek(self):
        if not self.stack_out:
            if not self.stack_in:
                raise IndexError("peek from empty queue")
            while self.stack_in:
                self.stack_out.append(self.stack_in.pop())
        return self.stack_out[-1]

    def is_empty(self):
        return not self.stack_in and not self.stack_out
```
